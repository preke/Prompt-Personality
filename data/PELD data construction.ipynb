```python
import pandas as pd
import numpy as np


PELD = pd.read_csv('Dyadic_PELD.tsv', sep='\t')
print(PELD['Speaker_1'].unique())
PELD.head()
```

```python
df_PELD = PELD[["Speaker_1","Speaker_2", "Utterance_1", "Utterance_2", "Utterance_3", \
                "Emotion_1", "Emotion_2", "Emotion_3"]]

def get_data(df, speaker):
    df_tmp = df
    df_tmp['dialog_state'] = df_tmp['Speaker_1'].apply(lambda x: [1,0,1])
    df_tmp['labels'] = df_tmp['Speaker_1'].apply(lambda x: 1 if speaker == x else 0)
    df_tmp['utterance'] = df_tmp.apply(lambda r: r['Utterance_1'] + ' ' + r['Utterance_3'], axis=1)
    df_tmp['context'] = df_tmp['Utterance_2']
    df_tmp['sent'] = df_tmp.apply(lambda r: r['Utterance_1'] + ' ' + r['Utterance_2']  + ' ' + r['Utterance_3'], axis=1)
    df_tmp['emo_label'] = df_tmp.apply(lambda r: [r['Emotion_1'], r['Emotion_2'], r['Emotion_3']], axis=1)
    df_tmp['raw_text']= df_tmp.apply(lambda r: [[r['Speaker_1'], r['Utterance_1']], \
                                                [r['Speaker_2'], r['Utterance_2']], \
                                                [r['Speaker_1'], r['Utterance_3']]], axis = 1)
    print(sum(df_tmp['labels'])/len(df_tmp))
    # df_tmp.to_csv('PELD_'+speaker+'.tsv', sep='\t')
    

get_data(df_PELD, speaker='Chandler')
get_data(df_PELD, speaker='Joey')
get_data(df_PELD, speaker='Rachel')
get_data(df_PELD, speaker='Monica')
get_data(df_PELD, speaker='Phoebe')
get_data(df_PELD, speaker='Ross')
```
