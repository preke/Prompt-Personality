```python
df_test = pd.read_csv('EAR_script.tsv', sep='\t')


df_test['ext.s'] = df_test['ext.s'].apply(lambda x: x/5.0)
df_test['ems.s'] = df_test['ems.s'].apply(lambda x: x/5.0)
df_test['ope.s'] = df_test['ope.s'].apply(lambda x: x/5.0)
df_test['agr.s'] = df_test['agr.s'].apply(lambda x: x/5.0)
df_test['con.s'] = df_test['con.s'].apply(lambda x: x/5.0)


df_test['ext.cls'] = df_test['ext.s'].apply(lambda x: '1' if x > 0.737500 else '0')
df_test['ems.cls'] = df_test['ems.s'].apply(lambda x: '1' if x > 0.700000 else '0')
df_test['ope.cls'] = df_test['ope.s'].apply(lambda x: '1' if x > 0.780000 else '0')
df_test['agr.cls'] = df_test['agr.s'].apply(lambda x: '1' if x > 0.777778 else '0')
df_test['con.cls'] = df_test['con.s'].apply(lambda x: '1' if x > 0.688889 else '0')

df_test.head()
```
