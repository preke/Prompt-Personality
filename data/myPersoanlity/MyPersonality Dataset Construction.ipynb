```python
myPersonalityGroup = myPersonality.groupby('#AUTHID').agg({'utterance':lambda x:''.join(x), 'cAGR':'min', 'cCON':'min', 'cEXT':'min', 'cOPN':'min', 'cNEU':'min'})
myPersonalityGroup.reset_index(inplace=True)
myPersonalityGroup
# x.to_csv('x_whole.tsv',sep = '\t')
```

```python
# show personality counts
df_labels = myPersonalityGroup[['cAGR','cCON','cEXT','cOPN','cNEU']]
df_counts = pd.DataFrame([])
df_counts['AGR'] = df_labels['cAGR'].value_counts().reset_index()['cAGR']
df_counts['CON'] = df_labels['cCON'].value_counts().reset_index()['cCON']
df_counts['EXT'] = df_labels['cEXT'].value_counts().reset_index()['cEXT']
df_counts['OPN'] = df_labels['cOPN'].value_counts().reset_index()['cOPN']
df_counts['NEU'] = df_labels['cNEU'].value_counts().reset_index()['cNEU']
print(df_counts)
```

```python
print((df_counts[0:1]/df_counts.sum()).round(decimals=3))
print((df_counts[1:2]/df_counts.sum()).round(decimals=3))
```

```python
# plot
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize # import figsize

import seaborn as sns

df_counts.index = (['False', 'True'])
ax = df_counts.T.plot.bar(stacked=True, color=['#FC7459', '#DBEAFF'])
# ax.color_palette("pastel")
# annotate the bars
for i, rect in enumerate(ax.patches):
    # Find where everything is located
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    y = rect.get_y()

    # The height of the bar is the count value and can used as the label
    label_text = f'{height:.0f}'

    label_x = x + width / 2
    label_y = y + height / 2

    # don't include label if it's equivalently 0
    if height > 0.001:
        ax.text(label_x, label_y, label_text, ha='center', va='center', fontsize=12)

ax.set_xticklabels(df_counts.columns, rotation=23, fontsize=12, fontstyle='oblique', fontweight='550')
ax.legend(bbox_to_anchor=(0.75, 0.99), loc=0, borderaxespad=0.0)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(True)
ax.spines['bottom'].set_visible(True)
plt.savefig('plot/myPersonality_personality.png', dpi=300)
```

```python
myPersonalityGroup['sample_length'] = myPersonalityGroup['utterance'].apply(lambda x: len(x.split(' ')))
myPersonalityGroup['sample_length'].describe()
```

```python
# split data and output tsv
labels_map = {'n':0,'y':1}
for index, row in myPersonalityGroup.iteritems():
    if index == 'cEXT':
        myPersonalityGroup_E = myPersonalityGroup[['utterance','cEXT']]
        myPersonalityGroup_E.rename(columns={'cEXT':'labels'},inplace=True)
        myPersonalityGroup_E['labels'] = myPersonalityGroup_E['labels'].map(labels_map)
        myPersonalityGroup_E.to_csv('MyPersonality_E_whole.tsv',sep = '\t')
    elif index == 'cNEU':
        myPersonalityGroup_N = myPersonalityGroup[['utterance','cNEU']]
        myPersonalityGroup_N.rename(columns={'cNEU':'labels'},inplace=True)
        myPersonalityGroup_N['labels'] = myPersonalityGroup_N['labels'].map(labels_map)
        myPersonalityGroup_N.to_csv('MyPersonality_N_whole.tsv',sep = '\t')
    elif index == 'cAGR':
        myPersonalityGroup_A = myPersonalityGroup[['utterance','cAGR']]
        myPersonalityGroup_A.rename(columns={'cAGR':'labels'},inplace=True)
        myPersonalityGroup_A['labels'] = myPersonalityGroup_A['labels'].map(labels_map)
        myPersonalityGroup_A.to_csv('MyPersonality_A_whole.tsv',sep = '\t')
    elif index == 'cCON':
        myPersonalityGroup_C = myPersonalityGroup[['utterance','cCON']]
        myPersonalityGroup_C.rename(columns={'cCON':'labels'},inplace=True)
        myPersonalityGroup_C['labels'] = myPersonalityGroup_C['labels'].map(labels_map)
        myPersonalityGroup_C.to_csv('MyPersonality_C_whole.tsv',sep = '\t')
    elif index == 'cOPN':
        myPersonalityGroup_O = myPersonalityGroup[['utterance','cOPN']]
        myPersonalityGroup_O.rename(columns={'cOPN':'labels'},inplace=True)
        myPersonalityGroup_O['labels'] = myPersonalityGroup_O['labels'].map(labels_map)
        myPersonalityGroup_O.to_csv('MyPersonality_O_whole.tsv',sep = '\t')
```
